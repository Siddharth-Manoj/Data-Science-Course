1.	Create a list of 5 integers. Perform and print the result of the following operations: append, extend, insert, remove, pop, clear, index, count, sort, and reverse.

```python
list = [10, 20, 30, 40, 50]
print(f"List: {list}")

#append - adding element to end (60)
list.append(60)
print(f"List after append: {list}")

#extend - add multiple elements to the end
list.extend([70, 80])
print(f"List after extend: {list}")

#insert - inserting element at a specific positon
list.insert(2, 25)
print(f"List after insert: {list}")

#remove - removing a aspecifc element
list.remove(20)
print(f"List after remove: {list}")

#pop - remove and return element at index
popped = list.pop(3)
print(f"After pop(3): {list}")

#clear - remove all elements
list1 = list.copy()
list1.clear()
print(f"After clear: {list1}")

#index - find index of first occurence
index_of_50 = list.index(50)
print(f"Index of 50 : {index_of_50}")

#count - count occureneces of a value
list.append(10)
count_of_10 = list.count(10)
print(f"Count of 10: {count_of_10}")

#sort - sort list in ascending order
list.sort()
print(f"After sort: {list}")

#reverse - reverse the list order
list.reverse()
print(f"After reverse: {list}")
```

11.	Create a small DataFrame manually with 10 rows and columns: 'Name', 'Age', 'City', and 'Salary'. Then:
- Use .info(), .describe()
- Select 'Name' and 'City' columns
- Drop 'City' column
- Fill any missing values in 'Salary' column with the mean
- Remove any duplicate rows


```python
import pandas as pd
import numpy as np


df = pd.DataFrame({
    'Name': ['Rahul','Aman','Ankit','Siddharth','Aditya','Divyanshu','Hemant','Ravi','Neha','Anjali'],
    'Age': [28, 25, 23, 24 ,27, 29, 30, 35, 23, 22],
    'City': ['Delhi', 'Mumbai', 'Bangalore', 'Chennai', 'Delhi', 'Chennai', 'Mumbai', 'Bangalore', 'Delhi', 'Bangalore'],
    'Salary': [45000, 52000, 68000, np.nan, 51000, 48000, 75000, 54000, 59000, 47000]
})

print(df)
print()

#info to show structures
print(df.info())
print()

#describe shows summary statistics
print(df.describe())
print()

#select specific columns
print(df[['Name', 'City']])
print()

#drop a column
df.drop('City', axis = 1, inplace = True)
print(df)
print()

#filling missing data with mean
mean_salary = df['Salary'].mean()
df['Salary'].fillna(mean_salary, inplace = True)
print(df)
print()

#remove duplicates if any
df.drop_duplicates(inplace = True)
print(df)
```
