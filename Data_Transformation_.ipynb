```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import IsolationForest

import ppscore as pps
```

```python
df = pd.read_csv("adult_with_headers (1).csv")

print(df.head())
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
# Replace ? with NaN
df.replace(" ?", np.nan, inplace=True)

# Check missing values again
print(df.isnull().sum())
```

```python
# Fill missing values using mode
df['workclass'].fillna(df['workclass'].mode()[0], inplace=True)
df['occupation'].fillna(df['occupation'].mode()[0], inplace=True)
df['native_country'].fillna(df['native_country'].mode()[0], inplace=True)

print(df.isnull().sum())
```

```python
# Numerical columns
num_cols = df.select_dtypes(include=['int64']).columns

# Categorical columns
cat_cols = df.select_dtypes(include=['object']).columns

print("Numerical Columns:")
print(num_cols)

print("Categorical Columns:")
print(cat_cols)
```

```python
""" STANDARD SCALING
standard scaling transforms the data so that mean becomes 0 standard deviation becomes 1

It is preferred when:
 - Data follows normal distribution
 - Algorithms like Logistic Regression, SVM, and PCA are used
 - Features have different ranges and need normalization

 It improves model performance and removes scale differences between features"""
```

```python
minmax_scaler = MinMaxScaler()

df_minmax = df.copy()
df_minmax[num_cols] = minmax_scaler.fit_transform(df[num_cols])

print(df_minmax.head())
```

```python
""" MIN MAX SCALING
it transforms the data into a fixed range between 0 and 1

it is preferred when neural networks and deep learning models are used
data should remain within a bounded range and features have different scales """

```

```python
# low-cardinality columns
low_cardinality = [col for col in cat_cols if df[col].nunique() < 5]

print(low_cardinality)
```

```python
df = pd.get_dummies(df, columns=low_cardinality, drop_first=True)

print(df.head())
```

```python
""" One Hot Encoding

it converts categorical values into binary columns

Pros
No ordinal relationship is introduced
Works well for low-cardinality features
Suitable for most machine learning algorithms
Prevents models from assuming numerical order

Cons
Increases the number of columns
May cause dimensionality issues for high-cardinality data
Training time may increase with many categories """
```

```python
#calculate categorical columns
cat_cols = df.select_dtypes(include=['object']).columns

print(cat_cols)
```

```python
le = LabelEncoder()

# Find high-cardinality columns
high_cardinality = [col for col in cat_cols if df[col].nunique() >= 5]

print(high_cardinality)
```

```python
# Apply Label Encoding
for col in high_cardinality:
    df[col] = le.fit_transform(df[col])

print(df.head())
```

```python
""" LABEL ENCODING

it converts categorical text values into numeric labels each unique category is assigned with a numerical value.

pros
memory efficient, faster and easy to implement

cons
may create false ordinal relationships
models may incorrectly assume higher numbers have greater importance
not suitable for nominal categorical data in some algorithms  """
```

```python
#Convert Boolean Columns into Integer
df['sex_ Male'] = df['sex_ Male'].astype(int)
df['income_ >50K'] = df['income_ >50K'].astype(int)

print(df.dtypes)
```

Feature Engineering

```python
#Create Age Group Feature
df['age_group'] = pd.cut(df['age'], bins=[0,25,45,65,100], labels=['Young','Adult','Senior','Old'])

print(df[['age','age_group']].head())  #income often depends on age
```

```python
#Create Work Hours Category Feature
df['work_hours_category'] = pd.cut(df['hours_per_week'], bins=[0,25,40,60,100], labels=['Part-time','Normal','Overtime','Heavy'])

print(df[['hours_per_week','work_hours_category']].head())  #working hours influence salary predictions
```

Checking skewness - indentifies skewed numerical features

```python
print(df[num_cols].skew())
```

```python
""" Log transformation

capital_gain is highly skewed.
"""
```

```python
df['capital_gain_log'] = np.log1p(df['capital_gain'])

print(df[['capital_gain','capital_gain_log']].head())
```

Isolation Forest for Outlier Detection
 -  detects and removes outliers from the dataset

```python
iso = IsolationForest(contamination=0.05)
outliers = iso.fit_predict(df[num_cols])
df = df[outliers == 1]

print(df.shape)
```

PPS Score Analysis
 - identifies features that are most predictive of the target variable.

```python
pps_matrix = pps.matrix(df)

print(pps_matrix[['x','y','ppscore']])
```
