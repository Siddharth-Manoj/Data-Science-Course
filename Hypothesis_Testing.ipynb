Build 99% Confidence Interval Using Sample Standard Deviation

```python
import numpy as np
import scipy.stats as stats
```

```python
data = [1.13, 1.55, 1.43, 0.92, 1.25, 1.36, 1.32, 0.85, 1.07, 1.48, 1.20, 1.33, 1.18, 1.22, 1.29]
```

```python
sample = np.array(data)
```

```python
mean = np.mean(sample)
std_dev = np.std(sample, ddof=1)
n = len(sample)
```

```python
#standard error
sem = std_dev/np.sqrt(n)
```

```python
#99% CI using T distribution
confidence = 0.99
df = n - 1

ci = stats.t.interval(confidence, df, loc=mean, scale=sem)

print("Sample Mean:", mean)
print("Sample Std Dev:", std_dev)
print("99% Confidence Interval:", ci)
```

```python
"""
in first step we collected the sample data which was a random sample of 15 print heads

in second step we calcualted the average (mean) durability of the sample

in third step we calculate the sample standard deviation(std_dev) since the population variablity was unknown

in fourth we calculate the standard error SE = s/sqrt(n) sample size was 15

Reason for using t-distribution
1. small sample size(15) which is less that 30 hence T distribution will give more accurate estimate
2. unknown population variablity
"""
```

HYPOTHESIS TESTING



State the Hypothesis


```python
"""
H0 : Weekly operating cost is equal to model cost.
H1 : Weekly operating cost is higher than model cost.
"""
```

Calculate the Test Statistic:
Use the following formula to calculate the test statistic(z):
where:
●	ˉxˉ = sample mean weekly cost (Rs. 3,050)
●	μ = theoretical mean weekly cost according to the cost model (W = $1,000 + $5X for X = 600 units)
●	σ = 5*25 units
●	n = sample size (25 restaurants)


```python
from scipy.stats import norm
import math

x_bar = 3050
mu = 1000 + 5 * 600
sigma = 5 * 25
n = 25

z = (x_bar - mu) / (sigma / math.sqrt(n))

print("Test statistic =", z)
```

Determine the Probability and compare:
Using the alpha level of 5% (α = 0.05),


```python
critical_value = norm.ppf(0.95)

print("Critical value =", critical_value)
```

Make a Decision:
Compare the test statistic with the critical value to decide whether to reject the null hypothesis.


```python
if z > critical_value:
    print("Reject Null Hypothesis")
else:
    print("Fail to Reject Null Hypothesis")
```
